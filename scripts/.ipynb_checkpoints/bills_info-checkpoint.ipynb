```python
senators = pd.read_csv('senators114.csv', index_col=0)
```

```python
senators.pivot_table(index='state', values='missed_votes_pct').sort_values(ascending=False)[:10].plot(kind='bar')
plt.xticks(rotation=0)
plt.legend(['Missed Votes, %'])
plt.show()
```

```python
senators[['last_name', 'missed_votes']].sort_values(by='missed_votes', ascending=False)[:20].plot(kind='bar', x='last_name')
plt.xticks(rotation=90)
plt.title('Gone Missing: \n Missed Votes in Senate 114')
plt.legend('')
plt.xlabel('')
plt.style.use('fivethirtyeight')
plt.tight_layout()
plt.savefig('missed_senators.png')
```

```python
#seniority distribution by party
import seaborn as sns
```

```python
def plot_distribution( df , var , target , **kwargs ):
    row = kwargs.get( 'row' , None )
    col = kwargs.get( 'col' , None )
    facet = sns.FacetGrid( df , hue=target , aspect=4 , row = row , col = col )
    facet.map( sns.kdeplot , var , shade= True )
    facet.set( xlim=( df[var].min() , df[ var ].max() ) )
    facet.add_legend()
```

```python
plot_distribution(senators, 'seniority','party')
```

```python
plot_distribution(senators, 'missed_votes', 'party')
```

```python
plot_distribution(senators, 'missed_votes', 'next_election')
```

```python
senators.groupby(['party', 'next_election']).size().unstack().plot(kind='bar', stacked=False)
plt.style.use('fivethirtyeight')
plt.legend(loc=0)
plt.title('Senators up for Reelection')
plt.xlabel('Party')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('reelection.png')
```

```python
senators.groupby(['party', 'next_election'])['missed_votes'].mean().unstack().plot(kind='bar', stacked=False)
plt.legend(loc=0)
plt.title('Senate 114: \n Missed Votes by Party, Next Election Year')
plt.xlabel('Party')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('party_elec.png')
```

```python
def title_parser(title):
    if 'Cloture' in title: return 'Invoke Cloture'
    if 'Table' in title: return 'Table'
    if 'Confirmation' in title: return 'Confirmation'
    if 'Nomination' in title: return 'Nomination'
    if 'Amdt' == title.split()[1]: return 'Amendment'
    if 'Amdt.' == title.split()[1]: return 'Amendment'
    if 'Admt.' == title.split()[1]: return 'Amendment'
    if 'Instruct' in title: return 'Instruct'
    if 'Waive' in title: return 'Waive'
    if 'Concur' in title: return 'Concur'
    if 'H.R' in title: return 'House Resolution'
    if 'joint' in title: return 'Joint Resolution'
    if 'concurrent' in title: return 'Concurr'
    if 'Conference' in title: return 'Conference Report'
    if 'S.'in title: return 'Senate Bill'
```

```python
bills['parsed_title'] = bills.title.apply(title_parser)
```

```python
bills.parsed_title.value_counts().plot(kind='bar')
```
